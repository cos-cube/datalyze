```python
from bs4 import BeautifulSoup
import numpy as np
import pandas as pd
import requests, re
```

```python
URL = "https://en.wikipedia.org/wiki/"
continentList = {"asia":"List_of_Asian_countries_by_population",
             "africa":"List_of_African_countries_by_population",
             "north":"List_of_North_American_countries_by_population",
             "south":"List_of_South_American_countries_by_population",
             "europe":"List_of_European_countries_by_population",
             "oceania":"List_of_Oceanian_countries_by_population"}
```

```python
request = requests.get(URL+continentList["europe"])
soup = BeautifulSoup(request.content,"html5lib")
```

```python
table = soup.findAll("tbody")[0]
count = []
popu = []
for row in table.findAll("tr")[2:-1]:
  country = row.find("th").text
  col = row.findAll("td")
  pop = col[0].text.replace(",","")
  count.append(country)
  popu.append(pop)

europe = pd.DataFrame({"country":count,"population":popu})
```

```python
europe.to_csv("europe.csv")
```

```python
def getData(URL,continent_):
  sendURL = URL+continent_
  print(sendURL)
  r = requests.get(sendURL)
  soup = BeautifulSoup(r.content,"html5lib")
  table = soup.findAll('tbody')[0]

  # country list definitions
  pop = []
  country = []
  fraction_of_pop = []
  per_1000_pop = []

  for row in table.findAll("tr")[1:-1]:
    col = row.findAll('td')
    if len(col) == 0:
      pass
    else:
      #print(col)
      country.append(col[0].text)

      #convert pop into int, generate percent pop
      population_ = col[1].text.replace(",","")
      pop.append(int(population_))

  #generate fraction of pop for value estimation
  sum_population = sum(pop)

  for i in pop:
    fraction = i/sum_population
    per_1000 = int(1000*fraction)

    fraction_of_pop.append(fraction)
    per_1000_pop.append(per_1000)

#created the dataset
  newdf = pd.DataFrame({"country":country,"population":pop,
                      "fraction_of_population":fraction_of_pop,
                      "per_1000_people":per_1000_pop})
  return newdf
```

```python
africadf = getData(URL,continentList["africa"])
```

```python
africadf.to_csv("africa.csv")
```
